## Data Enrichment

In addition to the provided dataset, we also added and derived new variables in Excel to further support the analysis of our research question:
- **party2016:** We added the winning political party to the 2016 Presidential Election for each state. We are interested in understanding if political allegiance had any significant impact to the timeliness of legislative action and its impact to infection rate (source: [MIT Election Data and Science Lab](https://dataverse.harvard.edu/dataset.xhtml?persistentId=doi:10.7910/DVN/42MVDX))
- **party2016_votepercent:** Represent the percentage of total popular vote the winning party received. Again, we're interested in whether the magnitude of party allegiance had any statistically significant impact to response time as well as potential impact to infection rate (source: [MIT Election Data and Science Lab](https://dataverse.harvard.edu/dataset.xhtml?persistentId=doi:10.7910/DVN/42MVDX))
- **CasesInLast7Days_100k:** Since each state's population can vary widely, we control for population by calculating the number of positive cases in the last 7 days per 100,000 residents
- **totalTestResults_100k:** Since each state's population can vary widely, we control for population by calculating the number of test results in the last 7 days per 100,000 residents
- **infection rate:** Total number of positive cases divided by the total number of test results. This is also our dependent variable. We opted for this definition of infection rate because basing the infection rate off of each state's total population (population 2018 variable) does not account for changes in population between 2018 and 2020, testing capacity, the fact that not every resident was tested, and certain individuals account for multiple tests, which diminished the magnitude of infection rate.
- **soe_to_sip:** Number of days between the state of emergency was declared and shelter-in-place was ordered. We're interested in understanding whether the timeliness of stay at home orders following the the state of emergency had any significant impact to fatality rate
    - For the state of California, Kentucky, New Mexico and New York, their respective shelter-in-place order is still active. Thus, the value was derived by taking the difference between the shelter-in-place start date and 7/7/2020, which is the date of one of the sources used to compile the dataset (source: [COVID-19 US state policy database 07_07_2020](https://github.com/KristenNocka/COVID-19-US-State-Policy-Database))
- **sip_binary:** Binary version of stay at home/shelter-in-place variable; coded a "1" for states that had a shelter-in-place order and a "0" otherwise
- **sip_start_end:** Measures the number of days the shelter-in-place order was active (difference between start and rescind date). For states that did not have a shelter-in-place order, the value is 0
- **soe_biz_close:** Represents the number of days between a state of emergency was declared and the shutdown of all non-essential businesses. A few select states did not have such policy, so the value is appropriately 0
- **biz_close_binary:** Binary version of closed non-essential businesses variable; coded a "1" for states that an explicit order from the governor's office to close all non-essential businesses and a "0" otherwise. Even if a state had shut down certain elements of non-essential businesses (i.e. dine-in restaurants, gyms, fashion retailer) but did not have an executive order from the state to shut down **all** non-essential business, we still coded it as 0
- **biz_close_open:** The number of days between the closure of non-essential businesses and business re-open date. There are 10 states in which had a statewide business re-open date, but did not have an initial closure date. This begs the question of how can a state re-open its businesses when it was never shutdown to close in the first place? Upon further examination
- **soe_face_mask:** Number of days between the state of emergency declaration and when face masks became mandatory for employees in public-facing businesses
- **face_mask_binary:** Binary version of mandate face mask use by employees in public-facing businesses variable; coded a "1" for states that put forth such order and a "0" otherwise
- **all_deathrate18:** Percent of each state's 2018 population that died for all causes. We would like to assess if fatality rate in pre-Covid world can influence Covid induced deaths (i.e. Does a state with an unhealthy population are more susceptible in dying from Covid?)

### State Response and Policy Duration

In this section, we will examine our main independent variables of interest identify extreme values and determine the most appropriate treatment option: **soe_to_sip, sip_start_end, soe_biz_close, biz_close_open, and soe_face_mask** Specifically, we hypothesize:
- soe_to_sip: While declaring a state of emergency is the first act to recognize the seriousness of the pandemic, the declaration itself is a piece of paper and have minimal impact to people's behavior. Measuring the number of days before state legislators followed-up with actionable response such as shelter-in-place can potentially curb infection rate
- sip_start_end: Shelter-in-place reduces contact rate between individuals. While it's a blunt force instrument to curb overall human contacts, it's also effective at stopping the spread of respiratory diseases like Covid-19. Thus, a logical extrapolation is that the longer the duration of shelter-in-place, the less severe infection rate will be
- soe_biz_close: Although governments can issue shelter-in-place, most orders were not enforced by local law enforcement agencies. Thus, the cost of defying orders from a legal perspective was low. The closure of all non-essential businesses was another avenue to disincentivize residents from leaving their home.
- biz_close_open: Our belief is that the duration of business closure can impact infection rate; the sooner the closure is lifted, the higher the infection rate because it offers options for people to congregate outside of their homes, especially coupled with quarantine fatigue.
- soe_face_mask: Even as non-essential businesses were shutdown and residents were ordered to stay home, essential businesses stayed open. However, essential employees, along with the rest of the country, were not required to wear a face mask when going outside. ### Model Diagnostics

We now review 4 diagnostic plots as well as statistical tests to help us check on some of the Classical Linear Model (CLM) assumptions. We use a combination of both visualization and tests because while visual inspection is often a first step, it's imperfect in nature and sometimes require more precise instrumentation to determine adherence with the CLM assumptions. However, there are 3 CLT assumptions that's difficult to determine through visual inspection of diagnostic plots:
- **MLR.1 Linearity in Parameters:** For model 1, we assume linearity in parameters by default. We have not transformed our dependent variable infection_rate or any of the independent variables. Thus, MLR.1 holds.
- **MLR.2 Random Sampling:** We assume that MLR.2 holds, since the dataset represents observations of all 51 states, and it would be difficult to argue that the observations are not representative of the country.
- **MLR.3 No Perfect Collinearity:** We used the Variance Inflation Score (VIF) to determine if there are symptoms of multicollinearity among our independent variables (see VIF score below), which is not easily identifiable through the diagnostic plots.
>**1. Residuals vs. Fitted:** The plot reveals a slight non-linear pattern in the plot below. However, because the bend is slight in nature, we do not believe that it violates the zero conditional mean of errors. Thus, the assumption holds (**MLR.4**)
>
>**2. Normal Q-Q:** This plot shows if residuals are normally distributed if it follows a relatively straight line. From the plot below, the bulk of the error terms seem to follow the straight line which suggests a fairly normal distribution. Additionally, the Shapiro-Wilk normality test returned a p-value of 0.9295 (see test below plots), further supporting the evidence that the normality of errors/residuals assumption holds (**MLR.6**)
>
>**3. Scale-Location:** This plot shows if residuals are spread equally along the ranges of explanatory variables. From the plot below, the red line is downward sloping but is relatively linear with equal (random) spread of residual points. Thus, we believe the homoscedasticity (homogeneity of residuals' variance) assumption holds (**MLR.5**). This claim is also supported by the p-value of 0.0896 from the Breusch-Pagan test  (see test below plots).
>
>**4. Residuals vs. Leverage:** This plot helps us find influential or high leverage points if any. An influential value is a value can alter the results of the regression analysis and is associated with a large residual. In our specific plot below, all points are within the bounds of Cook's distance lines although observation 43 is nearing the edge.

```r
par(mfrow = c(2, 2), cex = 1.5)
plot(model1, cex.caption = 1.5, cex.axis = 1, cex.lab = 1, pch = 20, cex = 1.5, lwd = 3)
```

**Shapiro-Wilk normality test (MLR.6):** If p-value > 0.05, then it implies that the distribution of the data are not significantly different from the normal distribution.

```r
shapiro.test(model1$residuals)
```

**Breusch-Pagan test (MLR.5):** If p-value < 0.05, then it implies the null hypothesis of homoscedasticity should be rejected. 

```r
bptest(model1)
```

**Variance Inflation Factor (VIF) scores (MLR.3):** Quantifies the extent of correlation between one predictor and the other predictor variables in a regression model (how much the variance of a regression coefficient is inflated due to multicollinearity in the model). It's used for diagnosing collinearity and multicollinearity. The higher the value, the greater the correlation of the variable with other variables. Values of more than 5 are sometimes regarded as problematic.

From the output below, all of our predictor variables are well below the 5 threshold, so we can claim that there is no perfect collinearity (**MLR.3**) in model 1.

```r
car::vif(model1)
```

Based on our review of the CLM assumptions, we believe that model 1 is an unbiased estimator of infection rate. However, the model has a poor fit and does not have much explanatory power. Thus, we will add additional covariates that will improve model fit while preserving adherence to the CLM assumptions.

# 6. Omitted Variables Discussion

Even by including all of the relevant variables provided in the data set to the linear regression, the resulting model still lacked explanatory power or high model fit. As mentioned before, part of the reason could be due to influential omitted variables that are either not provided or difficult to derive from the dataset. Below are some of the omitted variables that might be important along with how their absence may affect our results.
1. **Adherence to Policy:** As discussed in the modeling process section of the analysis, the most wide spread policies in the United States to combat the spread of Covid-19 have largely revolved shelter-in-place and the closure of non-essential businesses. However, most states have not enforced such policies the violation of such policies does not carry criminal or civil penalties. This is contrary to other countries, such as China, where a more draconian set of measures were put into effect to quickly stop the spread of the disease. While the policies are meant to curtail human contact, especially in densely populated establishments where social distancing is difficult (i.e. restaurants, bars, gyms, sporting events), we have no data regarding on how closely it was followed. It's one thing to enact a policy, but it's another for it to be enforced. For most states, most shelter-in-place orders offenders received no penalty. Moreover, some businesses have ignored non-essential business closure order and continued to operate until they were forcibly shutdown by the local law enforcement. For these reasons, we believe that increased adherence to shelter-in-place and non-essential business closure would result in decreased infection rate, since individuals are less likely to come into contact with another. Referencing other countries such as China and Italy who prosecuted individuals who broke such policies, they were able to contain the infection rate in a matter of weeks and not 3-4 months in which we find ourselves in the United States. Thus, we believe such variable would be practically significant and possess a high magnitude in reducing infection rate. A proxy measure could be used to operationalize such metric is the change in user's geo-location; if someone moved around more frequently than the city's norm/benchmark, then we can get a directional indicator of how tightly that individual is adhering to social distancing policies.

2. **General Face Covering Mandate:** Although the dataset contained a variable on when face coverings for public-facing employees were required, it does not include any data on whether the state issued face covering requirement for all residents when going beyond one's domicile. While there are plenty of debates regarding the effectiveness of face coverings in mitigating the spread of Covid-19, there is [emerging evidence for the effectiveness of cloth face coverings](https://www.cdc.gov/coronavirus/2019-ncov/prevent-getting-sick/cloth-face-cover-guidance.html#recent-studies) that has been compiled by the CDC. We believe this to be an important variable given that adherence to social distancing policies is a large unknown. If social distancing policies are not being enforced to the full extent of the law, then requiring a face covering to go outside for all residents is another deterrent that we believe can reduce the infection rate. This is especially pronounced in the United States because during the initial stages of the pandemic, public health officials recommended not to wear a face covering. By the time the recommendation was reversed April, we've lost valuable time. Referencing Asian countries that have effectively suppressed the spread of Covid-19, South Korea, Vietnam and Singapore to name a few, face coverings were part of daily life and much more accepted in their culture than it is in the United States. For these reasons, we believe a general face covering mandate could significantly reduce infection rate not only due to the ability of the covering to reduce droplets, but the social stigma of not wearing one in public establishments such as the grocery store. We also hypothesize that the variable would also be practically significant given the face mask requirement has been a consistent signal of success in Asian countries that have stopped the rapid spread of Covid-19.

3. **Population by Race:** While we have the population data by age, we do not have population by racial and minority groups. There are a growing body of research and studies that indicate minority groups contracting and dying from Covid-19 at a disproportionate rate than whites. An [article](https://www.cdc.gov/coronavirus/2019-ncov/need-extra-precautions/racial-ethnic-minorities.html) published by the CDC provides evidence that "Long-standing systemic health and social inequities have put some members of racial and ethnic minority groups at increased risk of getting Covid-19 or experiencing severe illness, regardless of age." More specifically, age-adjusted hospitalization rates are highest among non-Hispanic American Indian or Alaska Native and non-Hispanic black persons. Moreover, black persons are 5 times more likely to be hospitalized than whites. One plausible social factor that could partially explain the racial divide is that minorities are more likely to have Medicaid or high-deductible plans and are less likely to have a primary care physician. Without a primary care provider, people who thought they were infected were also likely to be turned away from hospitals.Thus, people who doesn't have a relationship with their primary care provider were much less likely to get tested. Moreover, testing sites were put in affluent communities, or required a car, and testing kits were in short supply and any time there is a shortage of something, minorities are less likely to get it. For these reasons, we hypothesize that states with a higher proportion of minorities and people of color relative to whites would increase both infection and fatality rate. Due to the growing body of evidence that's emerging from public health institutions and academia, we believe racial variables would have a moderate to high effect size on infection rate and fatality rate. 

4. **Public Transit Usage:** According to this [article](https://www.wsj.com/articles/public-transit-use-is-associated-with-higher-coronavirus-death-rates-researchers-find-11593342001) by the Wall Street Journal, the use of public transits are associated with increased Covid-19 infection and fatality rate. This makes intuitive sense, since buses, subways and light rails are confined spaces loaded with numerous people that are traveling to and from different locations. This is especially problematic because not only do these transits increase the likelihood of transmission within the confines of the modality, it also enables the disease to go mobile, since infected individuals are traveling from point A to B, potentially infecting others along the way. Moreover, those are more likely to use public transportation even in a global pandemic are individuals of low economic means who cannot afford a car and have no other choice. Because these individuals are already at a higher risk level of infection due to their economic stature, the use of public transits not compounds their own risk, but also increases the odds of infection to others unknowingly. Because public transits can be hotbed for transmission, especially for large cities that are densely populated (i.e. New York, San Francisco, Seattle, Boston), we believe collecting and operationalizing the data should not be a monumental task. Indeed, most major cities are already collecting a massive amount of usage data from their public transit services, which we can use to compare usage rate before and after the state of emergency was declared and/or shelter-in-place was ordered. For these reasons, we believe that a decreased use of public transits can decrease the transmission and infection of Covid-19. We also hypothesize that this variable would elicit a medium to high effect size on infection rate conditioned on population density.

5. **Weekly Jobless Claims by State:** Although the original dataset contained a variable on the maximum weekly amount of unemployment insurance, it does not provide any signal on the magnitude of unemployment due to Covid-19. While some states declared the closure of non-essential businesses, some allowed certain sectors to stay open or operate at reduced capacity. This discrepancy coupled with different dates of closure meant unemployment rate can vary from state to state and even from county to county (i.e. California). Such data is collected by the [United States Department of Labor](https://oui.doleta.gov/unemploy/claims.asp) and available on its website. Our hypothesis is that while the maximum insurance amount is measure of potential payout, the weekly jobless claims measures the magnitude of unemployment. Furthermore, we believe that the magnitude of unemployment has a positive correlation and influence on infection rate. If the unemployment rate is high for a particular state, then it's more likely for the unemployed individuals to seek employment, especially in essential businesses, which also carries higher risk of contraction. If the state also issued an order to close to all non-essential businesses, we suspect that more individuals will be seeking a lower supply of available jobs, which can increase the likelihood of human contact and infection. In addition, because the demand for food and grocery delivery services have increased by multiple-folds due to Covid-19, unemployed individuals are more likely to seek these type of high-risk jobs in order to meet ends meet. The risk is so prevalent among these workers, that some have banded together to [protest](https://www.thecut.com/2020/05/whole-foods-amazon-mcdonalds-among-coronavirus-strikes.html) the lack of protection and hazard pay. For the aforementioned reasons, we hypothesize that increases in the weekly jobless claims will increase Covid-19 infection rate and estimate a medium effect size due an overall low supply of job openings. 

6. **County Level Re-Opening Dates:** Since the re-opening dates in the dataset only provide state level information, it does not provide details at the county level. We believe county level re-opening dates to be important data with respect to infection rate because certain states such as California have delegated the re-opening decision to each county conditioned that they meet certain thresholds established at the state level. We are also seeing that states that have re-opened too quickly (i.e. Florida and Texas) are seeing a re-emergence of positive case count at a pace that was not observed before. Moreover, county level policy data would also provide additional insights into the effectiveness and timeliness of such responses that we could not observe that at the state altitude. County level information would provide additional samples that can be used for modeling from a goodness of fit and explanatory power perspective. More importantly, county level data could provide policy makers a more precise set of tools to create personalized approaches to reduce the spread of Covid-19; a one size fits all approach does not consider the potential differences in demographics, population, racial diversity, and socio-economic statuses, which can impact infection rate at different levels (example of [California](https://www.latimes.com/projects/california-coronavirus-cases-tracking-outbreak/)). While it's difficult to justify the direction and magnitude of infection rate without analyzing county level data first, we can use the current spikes in Covid-19 cases across the country as states began to re-open as a benchmark. For this reason, we believe counties that re-open too soon (i.e. short duration in shelter-in-place or business closure), especially bars and restaurants, are more likely to see an increase in infection rate. Additionally, we also estimate the impact to be of medium effect/magnitude.
